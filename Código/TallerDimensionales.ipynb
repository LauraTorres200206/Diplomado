```python
import pandas as pd
import requests
import sqlite3
conn = sqlite3.connect('Educacion.db')
# Cargar datos 
api_url = "https://www.datos.gov.co/resource/nudc-7mev.json?$limit=14585"
response = requests.get(api_url)
data = response.json()
df_educacion= pd.DataFrame(data)
df_educacion
```

```python
cursor = conn.cursor()
def ejecutar_sql(query, conexion=conn):
  """
  Toma una consulta SQL, la ejecuta y devuelve el resultado
  en una tabla de pandas para una visualización clara.
  """
  df = pd.read_sql_query(query, conexion)
  # Usamos display() porque en Colab presenta las tablas de forma más elegante.
  display(df)
```

```python
ft = pd.DataFrame(df_educacion.isnull().sum()).reset_index()
ft.columns = ["Variable","Faltantes"]
ft = ft.sort_values(["Faltantes"],ascending=False)
ft.index = range(ft.shape[0])
ft["% Faltantes"] = ft["Faltantes"] * 100 / df_educacion.shape[0]
ft.head(41)
```

```python
Total = len(df_educacion)
Var10mas = (df_educacion.isna().sum(axis=1) > 10).sum()
print(f"Eliminarías {Var10mas} filas de {Total} ({round(Var10mas/Total*100, 2)}%)")
```

```python
df_educacion['municipio'] = df_educacion['municipio'].str.replace(',', '', regex=False)
df_educacion['departamento'] = df_educacion['departamento'].str.replace(',', '', regex=False)
df_educacion['departamento'] = df_educacion['departamento'].str.replace('.', '', regex=False)
df_educacion['etc'] = df_educacion['etc'].str.replace(',', '', regex=False)
df_educacion['c_digo_etc'] = df_educacion['c_digo_etc'].str.replace('.', '', regex=False)
df_educacion['poblaci_n_5_16'] = df_educacion['poblaci_n_5_16'].str.replace('.', '', regex=False)
df_educacion['poblaci_n_5_16'] = df_educacion['poblaci_n_5_16'].str.replace(',', '', regex=False)
```

Esta etapa se realizó para estandarizar los nombres y corregir inconsistencias en los valores textuales. Por ejemplo, la ciudad de Bogotá aparecía registrada de distintas formas como Bogota, D.C o Bogotá D.C., lo que podría generar duplicados o errores en las dimensiones. Por ello, se eliminaron signos de puntuación como comas y puntos para unificar los registros.

```python
ColumnasConvertir = df_educacion.columns[8:41]
df_educacion['poblaci_n_5_16'] = pd.to_numeric(df_educacion['poblaci_n_5_16'], errors='coerce')
for col in ColumnasConvertir:
    df_educacion[col] = pd.to_numeric(df_educacion[col], errors='coerce')
```

```python
ColumnsSTR = [
    'a_o', 'c_digo_municipio', 'municipio',
    'c_digo_departamento', 'departamento',
    'c_digo_etc', 'etc'
]

df_educacion.loc[:, ColumnsSTR] = df_educacion.loc[:, ColumnsSTR].replace(r'^\s*$', pd.NA, regex=True)
df_educacion.loc[:, ColumnsSTR] = df_educacion.loc[:, ColumnsSTR].astype(str)
```

```python
filas_incompletas = df_educacion[df_educacion.isna().sum(axis=1) > 10]
df_educacion = df_educacion[df_educacion.isna().sum(axis=1) <= 10]
```

```python
ft = pd.DataFrame(df_educacion.isnull().sum()).reset_index()
ft.columns = ["Variable","Faltantes"]
ft = ft.sort_values(["Faltantes"],ascending=False)
ft.index = range(ft.shape[0])
ft["% Faltantes"] = ft["Faltantes"] * 100 / df_educacion.shape[0]
ft.head(41)
```

Se eliminaron las filas que presentaban 10 o más variables con datos faltantes, ya que estas no aportan suficiente información útil para el análisis. Al revisar los registros eliminados, se identificó que la mayoría pertenecen a departamentos con baja densidad poblacional, como Guainía, Vaupés y Amazonas. En particular, el municipio de Papunaua aparece en varios años sin datos en la mayoría de las variables, lo que sugiere dificultades estructurales en la recolección o reporte de información educativa en esa zona. Esta decisión de limpieza busca conservar la calidad del análisis, manteniendo registros más completos y representativos.

```python
ColumnsImputar = [
    'deserci_n_transici_n', 'deserci_n_media', 'deserci_n_secundaria',
    'deserci_n_primaria', 'repitencia_transici_n', 'repitencia_secundaria',
    'reprobaci_n_media', 'repitencia_primaria', 'repitencia', 'repitencia_media',
    'deserci_n', 'reprobaci_n_secundaria', 'cobertura_bruta_media',
    'tasa_matriculaci_n_5_16', 'cobertura_neta', 'cobertura_bruta_transici_n',
    'reprobaci_n_primaria', 'cobertura_bruta_primaria', 'cobertura_neta_primaria',
    'reprobaci_n', 'aprobaci_n_transici_n', 'reprobaci_n_transici_n',
    'aprobaci_n_media', 'cobertura_bruta_secundaria', 'cobertura_neta_secundaria',
    'cobertura_neta_media', 'cobertura_bruta', 'cobertura_neta_transici_n',
    'aprobaci_n_secundaria'
]

for col in ColumnsImputar:
    mediana = df_educacion[col].median(skipna=True)
    df_educacion[col].fillna(mediana, inplace=True)
```

```python
columnas_omitir = ['tama_o_promedio_de_grupo', 'sedes_conectadas_a_internet']

Data = df_educacion.drop(columns=columnas_omitir)
```

Se decidió omitir las variables tamaño_promedio_de_grupo y sedes_conectadas_a_internet debido a que presentaban un alto porcentaje de valores faltantes (cerca del 50%) y no eran relevantes para los objetivos del análisis. Las demás variables, al tener un porcentaje bajo de valores faltantes, fueron imputadas utilizando la mediana como método de estimación.

# Dimensiones

```python
Dimmunicipio = Data[['c_digo_municipio', 'municipio']].drop_duplicates().reset_index(drop=True)
Dimmunicipio['id_municipio'] = Dimmunicipio.index + 1  
DimDepartamento = Data[['c_digo_departamento', 'departamento']].drop_duplicates().reset_index(drop=True)
DimDepartamento['id_Departamento'] = DimDepartamento.index + 1  
Dimetc = Data[['c_digo_etc', 'etc']].drop_duplicates().reset_index(drop=True)
Dimetc['id_etc'] = Dimetc.index + 1  
DimTiemp = Data[['a_o']].drop_duplicates().reset_index(drop=True)
DimTiemp['id_Tiempo'] = DimTiemp.index + 1  
```

Creamos cuatro tablas dimensionales: una para los municipios, otra para los departamentos, una para las ETC y una para el tiempo (años). En cada una eliminamos los duplicados y asignamos un identificador numérico único (llave subrogada). Estas tablas nos permitirán conectar la información de forma organizada en la tabla de hechos.

# Tabla de hechos 
3.


```python
df_hechos = Data.copy()
df_hechos = df_hechos.merge(Dimmunicipio, on=['c_digo_municipio', 'municipio'], how='left')
df_hechos = df_hechos.merge(DimDepartamento, on=['c_digo_departamento', 'departamento'], how='left')
df_hechos = df_hechos.merge(Dimetc, on=['c_digo_etc', 'etc'], how='left')
df_hechos = df_hechos.merge(DimTiemp, on=['a_o'], how='left')
```

```python
ColumnasFinales = df_hechos.columns[7:45].tolist()
TablaHechos = df_hechos[ColumnasFinales].copy()
```

# Consultas SQL

```python
df_hechos.to_sql("hechos", conn, if_exists="replace", index=False)
Dimmunicipio.to_sql("Dimmunicipio", conn, if_exists="replace", index=False)
DimDepartamento.to_sql("DimDepartamento", conn, if_exists="replace", index=False)
Dimetc.to_sql("Dimetc", conn, if_exists="replace", index=False)
DimTiemp.to_sql("DimTiemp", conn, if_exists="replace", index=False)
```

```python
consulta_1 = """
SELECT 
    dm.municipio,
    AVG(f.tasa_matriculaci_n_5_16) AS Promedio_escolaridad
FROM 
    hechos f
JOIN 
    Dimmunicipio dm ON f.id_municipio = dm.id_municipio
GROUP BY 
    dm.municipio
ORDER BY 
    Promedio_escolaridad DESC;
"""
ejecutar_sql(consulta_1)
```

```python
consulta_2 = """
SELECT 
    dm.municipio,
    AVG(f.aprobaci_n) AS promedio_aprobacion,
    AVG(f.reprobaci_n) AS promedio_reprobacion
FROM 
    hechos f
JOIN 
    Dimmunicipio dm ON f.id_municipio = dm.id_municipio
GROUP BY 
    dm.municipio
ORDER BY 
    promedio_aprobacion DESC;
"""
ejecutar_sql(consulta_2)
```

```python
consulta_3 = """
SELECT 
    dd.departamento,
    AVG(f.cobertura_neta) AS promedio_cobertura_neta,
    AVG(f.cobertura_bruta) AS promedio_cobertura_bruta
FROM 
    hechos f
JOIN 
    DimDepartamento dd ON f.id_Departamento = dd.id_Departamento
GROUP BY 
    dd.departamento
ORDER BY 
    promedio_cobertura_neta DESC;
"""
ejecutar_sql(consulta_3)
```
